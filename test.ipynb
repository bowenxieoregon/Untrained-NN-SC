```python
import torch
from models import *
import numpy as np

input_depth = 1
NET_TYPE = 'skip'
pad = 'zero'
upsample_mode = 'nearest'

in_size = (51,51)
#in_size = (101,101)
net = get_net(input_depth, NET_TYPE, pad, upsample_mode, n_channels=1, act_fun='LeakyReLU', skip_n33d=512, skip_n33u=512, skip_n11=9, num_scales=10, downsample_mode='stride')
total_params = sum(p.numel() for p in net.parameters())
print(total_params)

state_dict = net.state_dict() 
param_vector = torch.cat([param.flatten() for param in state_dict.values()])

print(torch.norm(param_vector).item())
```

```python
import matplotlib.pyplot as plt
import numpy as np 
datat= np.load("strong_6_50_clear.npz")
datat = datat["T"]
print(datat.shape)
data1 = np.load("recover.npy")

fig, axs = plt.subplots(1, 2, figsize=(14, 6))

sel = 48
vmin = min(10 * np.log10(data1[:,:,sel]).min(), 10 * np.log10(datat[sel,:,:]).min())
vmax = max(10 * np.log10(data1[:,:,sel]).max(), 10 * np.log10(datat[sel,:,:]).max())

# Contourf plot for Matrix A
contour1 = axs[0].contourf(10 * np.log10(data1[:,:,sel]),levels = 500, vmin = vmin, vmax = vmax, cmap="jet")
# fig.colorbar(contour1, ax=axs[0], orientation='vertical')
axs[0].set_title('Contour plot for recovery')

# Contourf plot for Matrix B
contour2 = axs[1].contourf(10 * np.log10(datat[sel,:,:]), levels = 500,  vmin = vmin, vmax = vmax, cmap="jet")
# fig.colorbar(contour2, ax=axs[1], orientation='vertical')
axs[1].set_title('Contour plot for truth')

plt.show()
plt.close()
```
